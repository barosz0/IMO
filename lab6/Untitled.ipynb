```python
import matplotlib.pyplot as plt
```

```python
def read_data(filename):
    with open(filename) as f:
        lines = f.readlines()

        lenght = []
        first_scores = []
        second_scores = []
        avg_first_scores = []
        avg_second_scores = []
        
        for line in lines:
            line = line.replace("\n", "").split(";")

            lenght.append(float(line[0]))
            first_scores.append(float(line[1]))
            second_scores.append(float(line[2]))
            avg_first_scores.append(float(line[3]))
            avg_second_scores.append(float(line[4]))
    
    return lenght, first_scores, second_scores, avg_first_scores, avg_second_scores
```

```python
lenght, first_scores, second_scores, avg_first_scores, avg_second_scores = read_data("Output/test_kroB200.txt")
```

```python
plt.plot(lenght, first_scores, ".")
plt.xlabel("lenght")
plt.ylabel("score")
plt.show()
```

```python
plt.plot(lenght, second_scores, ".")
plt.xlabel("lenght")
plt.ylabel("score")
plt.show()
```

```python
plt.plot(lenght, avg_first_scores, ".")
plt.xlabel("lenght")
plt.ylabel("score")
plt.show()
```

```python
plt.plot(lenght, avg_second_scores, ".")
plt.xlabel("lenght")
plt.ylabel("score")
plt.show()
```
